 Exploratory Data Analysis on the Iris Dataset

```python
# Importing Required Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
data= sns.load_dataset("iris")
```

```python
# view first few rows
data.head()
```

```python
# view last few rows
data.tail()
```

```python
# basic information
data.info()
```

```python
# summary statistics
data.describe()
```

```python
# checking for missing values
data.isnull().sum()
```

```python
# count of values in species
data["species"].value_counts()
```

Initial Observations:
- The dataset contains 150 rows and 5 columns.
- There are no missing values.
- 4 numerical features and 1 categorical column 

Histogram for Feature Distributions

```python
# Histogram for each feature
data.hist(figsize=(10,6), edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data["sepal_length"])
```

Boxplot for Outlier detection

```python
#boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=data, orient="h")
plt.title("Boxplot of Features")
plt.show()
```

Pairplot for Relationships Between Features

```python
sns.pairplot(data, hue="species")
```

```python
# Only numerical columns
corr = data.drop("species", axis=1).corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

Insights & Observations:

- Petal length and petal width are highly correlated.
- Setosa species is clearly separable based on petal features.
- The dataset is balanced across the three species.
- There are no missing values or extreme outliers.
